## String Built in Method

```python
name = "Md. Nazim Uddin Alam Thepo"

#Length
print(len(name))

# Fina a part in the string
print("Nazim" in name)

#lower case upper case conversion
new_name = name.lower()
print(new_name)

# First letter capitalize rest of the letter in small letter
another_name = name.capitalize()
print(another_name)

# kon index theke substring ta suru hoiche (surur dik theke index ta dibe)
index = new_name.find("nazim")
print(index)

#last er dik theke find korte chaile
string = "mahi mahi mahi mahi mahi mahi mahi"
last_index = string.rfind("mahi")
print(last_index)

#frequency count
count = string.count("mahi")
print(f"Number of time Mahi is : {count}")

# replace specific part or substring
string.replace("mahi","nazim")
print(string) # no change cause its imutable for change we need to put it anyother variable

new_name=string.replace("mahi","Nazim")
print(new_name)
```

## List

```python
list = [10,30,340,340,340,45,56,67]
print(list)

#appending a number
list.append(4545454545)
print(list)

# adding anything is specific position
list.insert(0,45)
print(list)

# Deleting any value As Like STACK
list.pop()
print(list)

# Delete specific value
list.remove(340)  # ekbar occurance kore delete kore dey ekadik occurance thakle seta ar delete kore na
print(list)

# for sortng
list.sort()
print(list)

list.sort(reverse=True)  # T capital letter
print(list)

#Reverse
list.reverse()
print(list)
```

## Stack ( The list work as a stack )

## Queue ( The list work as a QUEUE )

```python
queue =[]
queue.append(1)
queue.append(2)   # append as like as stack
queue.append(3)
queue.append(4)

#but remove
print(f"front remove: {queue.pop(0)}")
print(queue)
```
