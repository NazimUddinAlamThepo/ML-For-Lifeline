```python
from functools import reduce

with open("./sample_data/example.txt", "r") as file:

  # readlines() create a list with each line

  string = file.readlines()

  # Number of Lines
  number_of_lines = len(string)

  # Number of Words
  number_of_words= list(map(lambda x : len(x.split()) , string ))
  total_number_of_words = sum(number_of_words)
  print(number_of_words)

  # map diye hobe na karon map sob element er upore kaj kore alada alada vab sob gula elemnets i return kore abar

  # reduce ekta single number return kore so we need to use reduce

  another_total_number_os_words = reduce( lambda x,y : x+y , number_of_words)
  print(another_total_number_os_words)


  # total characters

  # Step-01 : Cleaning ( Delete New line)
  string = list(map(str.strip , string))

  # remove space
  string = list(map(lambda x : x.replace(" ",""),string))

  total_characters = list(map(lambda x : len(x) , string))
  total_characters = reduce(lambda x, y : x+y ,total_characters)




print(f"Number of line : {number_of_lines}")
print(f"Number of words : {number_of_words}")
print(f"Number of words : {total_number_of_words}")
print(f"Number of characters : {total_characters}")
```

```python
with open("./sample_data/counter.txt", "w") as file:
  file.write(f"Number of line : {number_of_lines} \n")
  file.write(f"Number of words : {total_number_of_words} \n")
  file.write(f"Number of characters : {total_characters} \n")

```

```python
# Create and write sample email addresses to a text file
with open("emails.txt", "w") as file:
    file.write("john.doe@example.com\n")
    file.write("JANE.SMITH@GMAIL.COM\n")
    file.write("michael_brown@outlook.com\n")
    file.write("emily.jones@YAHOO.COM\n")
    file.write("john.doe@example.com\n")          # repeated
    file.write("sarah.connor@gmail.com\n")
    file.write("JANE.SMITH@GMAIL.COM\n")          # repeated in uppercase
    file.write("robert.king@company.org\n")
    file.write("alex.taylor@GMAIL.com\n")
    file.write("MICHAEL_BROWN@OUTLOOK.COM\n")     # same as lowercase version
```

```python
#converting all emails to lowercase
with open("emails.txt", "r") as file:
  file.seek(0)
  emails_list = file.readlines()

  emails_list = list(map(lambda x: str.lower(x) , emails_list))
  print(emails_list)
  print(len(emails_list))
```
