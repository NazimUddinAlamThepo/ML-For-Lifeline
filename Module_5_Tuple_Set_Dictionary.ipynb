```python
# Dictionary View method
# Real Time kaj kore

dipu = {
    "name": "Mahi Alam Dipu",
    "id": "2003099",
    "department": "Computer Science",
    "cgpa": 3.80,
    "courses": ["Compiler Design", "Data Science", "Networking"]
}

keys = dipu.keys() # It will change dynamically
values = dipu.values()

dipu["result"] ="Good"

print(keys)
print(values)

# Unpacking to iterate)

for key,value in dipu :   # too many values to unpack (expected 2)
  print(key, value)
```

```python
# For solving the issue of unpacking we will use item
# Real Time Update hoy (Update kora lage na just pore oita niye kaj korlei oita updated value ta niye kaj kore)

item = dipu.items()

# all the key, value are in tuple () format

print(item)

for key , value in dipu.items() :  # It can be unpacked
  print(key,value)
```

```python
# All about ZIP


coordinates = [
    (23.8103, 90.4125),   # Dhaka
    (22.3569, 91.7832),   # Chittagong
    (24.8949, 91.8687),   # Sylhet
    (22.8456, 89.5403),   # Khulna
    (24.3740, 88.6010),   # Rajshahi
    (22.7010, 90.3535),   # Barishal
    (25.7439, 89.2752),   # Rangpur
    (21.4272, 92.0058)    # Cox’s Bazar
]


places = [
    "Dhaka",
    "Chittagong",
    "Sylhet",
    "Khulna",
    "Rajshahi",
    "Barishal",
    "Rangpur",
    "Cox’s Bazar"
]

exact_location = { X : Y for X , Y in
                  zip(places,coordinates)}

print(exact_location)
```

```python
# Syntax: list[start : stop : step]
# start = where to begin (inclusive)
# stop  = where to stop (exclusive)
# step  = how many elements to skip (+ for forward, - for backward)
# ==============================================================

# Create a list
lst = [10, 20, 30, 40, 50, 60]
print("Original List:", lst)

# Index Reference
print("\nINDEX REFERENCE:")
print("Positive Index:  0   1   2   3   4   5")
print("Negative Index: -6  -5  -4  -3  -2  -1")
print("List Elements:  ", lst)

# ==============================================================
# 1️⃣ BASIC FORWARD SLICING
# ==============================================================

print("\n1️⃣ lst[0:2] -> from index 0 up to (not including) index 2")
print("Output:", lst[0:2])  # [10, 20]

print("\n2️⃣ lst[-3:] -> last 3 elements starting from -3 to end")
print("Output:", lst[-3:])  # [40, 50, 60]

print("\n3️⃣ lst[-1:] -> only last element")
print("Output:", lst[-1:])  # [60]

print("\n4️⃣ lst[2:] -> from index 2 to the end")
print("Output:", lst[2:])  # [30, 40, 50, 60]

# ==============================================================
# 2️⃣ SLICING WITH MISSING START OR STOP
# ==============================================================

print("\n5️⃣ lst[:-1] -> all elements except the last")
print("Output:", lst[:-1])  # [10, 20, 30, 40, 50]

print("\n6️⃣ lst[:1] -> only first element (up to index 1)")
print("Output:", lst[:1])  # [10]

print("\n7️⃣ lst[:2] -> first two elements")
print("Output:", lst[:2])  # [10, 20]

print("\n8️⃣ lst[-4:-1] -> from -4 to before -1")
print("Output:", lst[-4:-1])  # [30, 40, 50]

# ==============================================================
# 3️⃣ SLICING WITH STEP VALUES
# ==============================================================

print("\n9️⃣ lst[::2] -> every 2nd element (forward direction)")
print("Output:", lst[::2])  # [10, 30, 50]

print("\n🔟 lst[::-2] -> every 2nd element (backward direction)")
print("Output:", lst[::-2])  # [60, 40, 20]

print("\n1️⃣1️⃣ lst[2::] -> from index 2 to end (default step = 1)")
print("Output:", lst[2::])  # [30, 40, 50, 60]

print("\n1️⃣2️⃣ lst[-2::] ")

print("Output:", lst[-2::])  # [30, 40, 50, 60]

# ==============================================================
# 4️⃣ REVERSE SLICING
# ==============================================================

print("\n1️⃣3️⃣ lst[::-1] -> reverse the entire list")
print("Output:", lst[::-1])  # [60, 50, 40, 30, 20, 10]
print("Explanation: step = -1 reverses the direction of traversal.")



```
